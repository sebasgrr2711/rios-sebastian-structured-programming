### Key Concepts:
- **`open()` function**: The primary way to open a file.
- **Modes**: `'r'` (read), `'w'` (write), `'a'` (append).
- **Context Managers (`with` statement)**: The best practice to ensure files are closed automatically.

## 1. Writing vs Overwriting

When you open a file in `'w'` mode, Python creates the file if it doesn't exist. **Warning:** If the file already exists, `'w'` will delete everything inside it before writing new data.

```python
# Creating a fresh file
with open('students.txt', 'w', encoding='utf-8') as f:
    f.write('Alice\nBob\nCharlie')

print("File 'students.txt' created with initial names.")
```

## 2. Iterating Over Lines

For large files, it is inefficient to use `.read()`. Instead, we can loop over the file object to process one line at a time. This saves memory.

```python
print("Reading names from the file:")
with open('students.txt', 'r', encoding='utf-8') as f:
    for line in f:
        # .strip() removes the newline character '\n'
        print(f"Student: {line.strip()}")
```

## 3. Appending to a File

Use the `'a'` mode to add content to the end of an existing file without deleting its previous content.

```python
with open('ejemplo.txt', 'a', encoding='utf-8') as f:
    f.write('\nThis line was appended later!')

with open('ejemplo.txt', 'r') as f:
    print(f.read())
```

### Exercise 2: Word Counter
Read the file `ejemplo.txt` and count how many words are in it. (Hint: use `.split()` on the string).

```python
# Exercise 2 Solution
try:
    with open('ejemplo.txt', 'r') as f:
        content = f.read()
        words = content.split()
        print(f"The file contains {len(words)} words.")
except FileNotFoundError:
    print("Please run the first cells of the notebook to create the file first.")
```

### Exercise 3: File Copy Tool
Write a program that reads the content of `students.txt` and writes it into a new file called `students_backup.txt`. This is a common real-world task.

```python
carrera=input("Carrera")
with open("students.txt", "w" )as source:
    source.write(carrera)
with open("students.txt", "r") as source:
    content = source.read()


with open("students_backup.txt", "w") as backup:
    backup.write(content)

with open("students.txt", "r") as backup:
    print(backup.read())
print("Backup created successfully!")
```
